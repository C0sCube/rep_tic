```python
import os
import logging
from datetime import datetime
import shutil

from app.utils import Helper
from app.logger import setup_logger
from app.market_parser import FutureOptionsWriter
from app.constants import OUTPUT_DIR,LOG_DIR,INPUT_DIR,DATA_DIR,CONFIG, conf_sp

def sorted_rate_files(path):
    files = os.listdir(path)
    rate_files = [f for f in files if f.endswith(".txt")]
    rate_files.sort(key=lambda x: int(x.split("_")[1].split(".")[0]))
    return rate_files

if __name__ == "__main__":

    utils = Helper()
    logger = setup_logger("sample",LOG_DIR,set_global=True)
    config = CONFIG
    writer = FutureOptionsWriter(
        logger=logger,
        config=config,
        output_path=OUTPUT_DIR,
        bin_size=100
    )

    print(OUTPUT_DIR)
    
    INPUT_DIR = r"C:\Users\kaustubh.keny\Downloads\20260417"
    
    logger = setup_logger("sample",LOG_DIR,set_global=True)

    dt_file = sorted_rate_files(INPUT_DIR)
    tick_count = 0
    for file_name in dt_file:
        file_path = os.path.join(INPUT_DIR, file_name)
        logger.info(f"processing: {file_name}")
        #file specific error handler
        try:
            lines = utils.read_file(file_path).splitlines()
            for idx,line in enumerate(lines):
                #ticker specific error handler
                try:
                    writer.process_ticker(line)
                except Exception as e:
                    logger.error(f"Ticker error in {file_name}: {type(e).__name__}: {e}")

                            
            tick_count += len(lines)
            # logger.info(f"Completed: {file_name} | Ticks: {len(lines)} | Total: {tick_count}")
        except Exception:
            raise

    writer.flush_all_data()

    print("\n Files written under:", OUTPUT_DIR)
```

```python
# --- STOCKS VALIDATION ---

file_path = "NSECME.txt"

total = 0
valid = 0
bad = 0
duplicates = 0

seen = set()

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        total += 1
        line = line.strip()

        # basic format check
        if not line.startswith("T||"):
            bad += 1
            continue

        valid += 1

        # duplicate detection (exact same tick)
        if line in seen:
            duplicates += 1
        else:
            seen.add(line)

print("---- STOCKS ----")
print("Total lines     :", total)
print("Valid ticks     :", valid)
print("Malformed lines :", bad)
print("Exact duplicates:", duplicates)

assert total == valid + bad, "Mismatch in counts!"
```
