```python
import os
from collections import Counter

folder_path = r"C:\Users\kaustubh.keny\Projects\OFFICE PROJECTS\PROG_NFO_WRITER\input"   # folder containing .txt files
code_sets = Counter()

for filename in os.listdir(folder_path):
    if not filename.endswith(".txt"):
        continue

    path = os.path.join(folder_path, filename)

    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            if "CPR" not in line:
                continue

            _, *fields = line.strip().split(";")
            if not fields:
                continue

            codes = [
                item.split("=")[0]
                for item in fields
                if "=" in item
            ]

            if codes:
                code_sets[frozenset(codes)] += 1


# ---- OUTPUT ----
print(f"Total unique code-sets: {len(code_sets)}\n")

for codeset, count in code_sets.items():
    print(f"{count} times -> {sorted(codeset)}")
```

```python
import os

def split_file_by_size(
    input_file,
    output_dir,
    chunk_size_mb=100,
    prefix="chunk"
):
    os.makedirs(output_dir, exist_ok=True)

    chunk_size = chunk_size_mb * 1024 * 1024
    chunk_index = 1
    current_size = 0

    out_file = None

    with open(input_file, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            if out_file is None or current_size >= chunk_size:
                if out_file:
                    out_file.close()

                output_path = os.path.join(
                    output_dir,
                    f"{prefix}_{chunk_index}.txt"
                )
                out_file = open(output_path, "w", encoding="utf-8")
                chunk_index += 1
                current_size = 0

            out_file.write(line)
            current_size += len(line.encode("utf-8"))

    if out_file:
        out_file.close()

split_file_by_size(
    input_file=r"C:\Users\kaustubh.keny\Downloads\SGDataOut7-10-2025.txt",
    output_dir="SG_chunks",
    chunk_size_mb=100
)
```

```python
# --- STOCKS VALIDATION ---

file_path = "NSECME.txt"

total = 0
valid = 0
bad = 0
duplicates = 0

seen = set()

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        total += 1
        line = line.strip()

        # basic format check
        if not line.startswith("T||"):
            bad += 1
            continue

        valid += 1

        # duplicate detection (exact same tick)
        if line in seen:
            duplicates += 1
        else:
            seen.add(line)

print("---- STOCKS ----")
print("Total lines     :", total)
print("Valid ticks     :", valid)
print("Malformed lines :", bad)
print("Exact duplicates:", duplicates)

assert total == valid + bad, "Mismatch in counts!"
```
